# Defense of the Ancient (DOTA)

# Introduction about the Game

This notebook is a mini game inspired by the game of Dota 2, one of the best eSports in the world with the highest prize pool of $31M as of 2019. 

The game is a multiplayer online battle arena game, meaning it is a 5v5 match between players. The two teams are call Radiant and Dire. Each player can select 1 hero to play in a 100+ heroes pool, each with different abilities, strengths and weaknesses. The game has a fixed map divided into three lanes players often called Mid, Bot and Top. The goal of the game is to destroy the opponent's Throne, a building centered in each team's base surround by 3 raxes at the end of the 3 lanes. The Throne is protected by Towers and the players. Players' heroes will match each other in lane and later on try to fight each other in order to complete the objectives.

This mini game tries to replicate a typical Dota match. First of all, the players will select their heroes to complete the team's draft, then assign roles and lanes to each hero. The draft phase is highly strategic as heroes do counter each other and heroes combination will give strategic advantage in the game. After hero selection comes the real game where the heroes are assigned in lane and role. They will face off each other. In this minigame, the heroes' early game matchup will give them an advantage before entering the middle stage of the game. This is important because an early advantage and some luck will give the team a unnstoppable force to win over the opponent (snowball). As mentioned earlier, each hero is unique and has their unique powers. Some heroes are really strong in the early game, some need time and resource in order to become powerful. As the result, the selection of hero and the strategy used in the game is what Dota focuses on. 

In the beginning second stage of the game (midgame), each team can decide the strategy they want to do in the game. There are 3 options: offensive (aggressive), defensive (turtle) or balance (normal). Aggressive strategy will give the team an immediate power boost to the team will have a stronger advantage prior to combat. However, the reward for this strategy is not as high as than the other two. Turtle strategy will not give a team an immediate power boost, but rather reward them greater if they win combat. If they lose combat, they will have receive any reward. The normal strategy is the balance between the two. A good example is if a team has a strong lategame heroes, the team can apply the turtle strategy. One win combat with the turtle strategy might reward more than one combat had the team chosen the other strategies. Vice versa, a team with strong early game knowing that they can't overcome the other team in the lategame, should go aggressive early on to maintain their snowball until the end of the game.  

In the midgame stage, I created series of combat which randomly assign the outcome of the combat. Each team will have 33% chance of winning, the other 33% will be a tie. An early advantage will give a team a greater chance of winning. For example, thanks to a strong early game and aggressive strategy, Radiant team chance of winning the first combat increases from 33% to 45%. The tie probability will always be constant of 33%. If Radiant team wins, they will gain greater power boost to the team than the Dire team. The winning team will also receive random bonus power boost. There are 6 random options. In the event of a tie, both teams will receive the same boost growth, but neither will receive a random boost.  

The number of combat is a random number from 5 to 10. At the end of the midgame/ combat stage, the teams difference in power will be taken into consideration in the first push in the lategame stage. In addition, if a team wins a certain number of combats in the midgame stage, the opponent will lose a certain number of rax or even Throne, causing the game to end here. Losing 3 rax will result Megacreep, which gives a significant advantage to the winning team. In the lategame stage there's no tie event anymore, each team by default will have 50% of winning, adjusted by the power difference and the number of rax left at the end of midgame stage. For example, at the end of combat stage, Radiant has 3 rax left but Dire only has 1 left, plus Radiant heroes are more powerful. This causes the winning push probability of Radiant from 50 to 85%. Radiant can either choose to push another rax to further its advantage or end the game by pushing Throne. Either option will have a probability of Dire making a comeback, winning and gaining even more growth. Remember at this stage some heroes have become either already too powerful or they grow outgrow everyone else. The game ends when a team successfully destroy the Throne.
